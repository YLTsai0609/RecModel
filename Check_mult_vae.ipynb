```python
# import scipy.sparse
import torch.optim
from RecModel import Mult_VAE
from sklearn.preprocessing import normalize
import numpy as np

train_mat = scipy.sparse.load_npz('data/mat_bin_train.npz')
eval_mat = scipy.sparse.load_npz('data/mat_bin_validate.npz')
#train_mat_normalized = normalize(X=train_mat, norm='l2', axis=1)

# Create a train test VAE.
test = Mult_VAE(k = 200, num_items = train_mat.shape[1], dense_layers_encoder=[train_mat.shape[1], 600], dense_layers_decoder=[600, 200], batch_norm_encoder=True,
            batch_norm_decoder=True, dropout_rate_decoder=0.5, dropout_rate_encoder=0.5, beta=0.2)

test.set_optimizer(torch.optim.Adam(test.parameters(), lr=1e-3))
test.set_writer('logs/')
```

```python
test.train(X_train=train_mat, batch_size=1500, epochs=5, verbose=1)
```

```python
res = test.eval_topn(test_mat=eval_mat, batch_size=1000, topn=np.array([4, 10, 20]), rand_sampled =1000, random_state=None)
print(res)
```
